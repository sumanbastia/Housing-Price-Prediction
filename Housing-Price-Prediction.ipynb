```python
# Check for missing values in the dataset
missing_values = data_copy.isnull().sum()

# Print the columns with missing values and their total count
print("\nColumns with missing values:\n", missing_values[missing_values > 0])

missing = missing_values[missing_values > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

From the above bar plot we can see that there are several columns with missing values. We will handle them using following methods:
* For the column where missing value is too low(below 5%), we can remove the whole row.
* For the columns where missing values are high, we will simply remove the columns or impute the missing values relatively to its datatypes if datatype is numerical, we will impute the missing value as its median and if datatype is object type, we will impute the missing value as its mode to avoid the outliers. 

```python
# Dropping the columns with high missing values
data_copy = data_copy.drop([ 'Id', 'PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu'], axis=1)

# Here we drop the column bcoz there is one missing value
data_copy = data_copy.dropna(subset=['Electrical'])

# Imputing the rest of missing value by their median and mode
data_copy['LotFrontage'] = data_copy['LotFrontage'].fillna(data_copy['LotFrontage'].median())
data_copy['GarageType'] = data_copy['GarageType'].fillna(data_copy['GarageType'].mode()[0])
data_copy['GarageYrBlt'] = data_copy['GarageYrBlt'].fillna(data_copy['GarageYrBlt'].median())
data_copy['GarageFinish'] = data_copy['GarageFinish'].fillna(data_copy['GarageFinish'].mode()[0])
data_copy['GarageQual'] = data_copy['GarageQual'].fillna(data_copy['GarageQual'].mode()[0])
data_copy['GarageCond'] = data_copy['GarageCond'].fillna(data_copy['GarageCond'].mode()[0])
data_copy['BsmtQual'] = data_copy['BsmtQual'].fillna(data_copy['BsmtQual'].mode()[0])
data_copy['BsmtCond'] = data_copy['BsmtCond'].fillna(data_copy['BsmtCond'].mode()[0])
data_copy['BsmtExposure'] = data_copy['BsmtExposure'].fillna(data_copy['BsmtExposure'].mode()[0])
data_copy['BsmtFinType1'] = data_copy['BsmtFinType1'].fillna(data_copy['BsmtFinType1'].mode()[0])
data_copy['BsmtFinType2'] = data_copy['BsmtFinType2'].fillna(data_copy['BsmtFinType2'].mode()[0])
data_copy['MasVnrType'] = data_copy['MasVnrType'].fillna(data_copy['MasVnrType'].mode()[0])
data_copy['MasVnrArea'] = data_copy['MasVnrArea'].fillna(data_copy['MasVnrArea'].median())
```

# Outlier Treatment

Finally, let's identify any outliers in the dataset and remove them. We can do it by different methods like drawing box plot, scatter plot and do them manually or by using z-value or IQR method. Here we use z-value method. 

```python
import scipy.stats as stats
for i in cols_num:
    #find absolute value of z-score for each observation
    z = np.abs(stats.zscore(data_copy[[i]]))
#only keep rows in dataframe with all z-scores less than absolute value of 3 
data_copy = data_copy[(z<3).all(axis=1)]

#find how many rows are left in the dataframe 
data_copy.shape
```

We can't use any diagram or plot to remove outlier here bcoz of large number of features which will take huge amount of time. And here I applied IQR also which remove almost half of the row from our dataset. That's why I used z-value method here.

# Linear Regression

```python
data_copy.shape
```

```python
X = data_copy[sel_cols]
y = data_copy['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression(fit_intercept=True)
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)

# R squared value
r2_test=reg.score(X_test,y_test)
r2_train=reg.score(X_train,y_train)
print(r2_test,r2_train)
```

```python
LRAcc = r2_test
print("Linear Regression best accuracy: {:.2f}%".format(LRAcc*100))
```

```python
# plot for residual error
  
## setting plot style
plt.style.use('fivethirtyeight')
  
## plotting residual errors in training data
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train,
            color = "green", s = 10, label = 'Train data')
  
## plotting residual errors in test data
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test,
            color = "blue", s = 10, label = 'Test data')
  
## plotting line for zero residual error
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
  
## plotting legend
plt.legend(loc = 'upper right')
  
## plot title
plt.title("Residual errors")
  
## method call for showing the plot
plt.show()
```

* By feature scaling we may get better visualization of how errors are distributed.

# K Nearest Neighbors(KNN model)

```python
from sklearn.neighbors import KNeighborsRegressor

scoreListknn = []
for i in range(1,50):
    KN_reg = KNeighborsRegressor(n_neighbors = i)
    KN_reg.fit(X_train, y_train)
    scoreListknn.append(KN_reg.score(X_test, y_test))
    
plt.plot(range(1,50), scoreListknn)
plt.xticks(np.arange(1,50,4))
plt.xlabel("K value")
plt.ylabel("Score")
plt.show()
KNAcc = max(scoreListknn)
print("KNN best accuracy: {:.2f}%".format(KNAcc*100))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

scoreListDT = []
for i in range(2,100):
    DT_reg = DecisionTreeRegressor(max_leaf_nodes=i)
    DT_reg.fit(X_train, y_train)
    scoreListDT.append(DT_reg.score(X_test, y_test))
    
plt.plot(range(2,100), scoreListDT)
plt.xticks(np.arange(2,100,8))
plt.xlabel("Leaf")
plt.ylabel("Score")
plt.show()
DTAcc = max(scoreListDT)
print("Decision Tree Accuracy:  {:.2f}%".format(DTAcc*100))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

scoreListRF = []
for i in range(2,20):
    RF_reg = RandomForestRegressor(n_estimators = 1000, random_state = 1, max_leaf_nodes=i)
    RF_reg.fit(X_train, y_train)
    scoreListRF.append(RF_reg.score(X_test, y_test))
    
plt.plot(range(2,20), scoreListRF)
plt.xticks(np.arange(2,20,5))
plt.xlabel("RF Value")
plt.ylabel("Score")
plt.show()

RFAcc = max(scoreListRF)
print("Random Forest Accuracy:  {:.2f}%".format(RFAcc*100))
```

# Gradiant Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
GB_reg = GradientBoostingRegressor()
GB_reg.fit(X_train, y_train)

y_pred = GB_reg.predict(X_test)

# R squared value
r2_test=GB_reg.score(X_test,y_test)
r2_train=GB_reg.score(X_train,y_train)
print(r2_test,r2_train)
```

```python
GBAcc = r2_test
print("Gradiant Boosting accuracy: {:.2f}%".format(GBAcc*100))
```
